# 1. Initialize Environment & Download Dataset

*Goal: Ensure the EVTX attack sample dataset is present before running any other notebooks.*

This notebook handles two common scenarios:
- **Local clone** — the repo was cloned without `--recurse-submodules`, so the dataset submodule directory is empty
- **GitHub Codespace** — Codespaces do not automatically initialize submodules, so the dataset must be pulled explicitly

Run each cell in order. If the dataset is already present, the steps are safe to re-run.

## Get Dataset Path

The EVTX datasamples should live in `datasets/EVTX-ATTACK-SAMPLES` submodule.

```python
from pathlib import Path

_cwd = Path.cwd()
print(f"Current directory: {_cwd}")

repo_root = _cwd.parents[1]
dataset_path = repo_root / "datasets" / "EVTX-ATTACK-SAMPLES"
print(f"Dataset Path: {dataset_path}")
```

## Initialize the Dataset Submodule

The dataset is tracked as a git submodule. If it was not downloaded during the initial clone this cell will fetch it.

This runs `git submodule update --init --recursive` from the repo root, which works correctly for both a local clone and a Codespace.

```python
import subprocess

evtx_files = list(dataset_path.rglob("*.evtx"))

if evtx_files:
    print(f"Dataset already present — found {len(evtx_files)} .evtx files.")
else:
    print("Dataset not found or empty. Initializing submodule...")
    result = subprocess.run(
        ["git", "submodule", "update", "--init", "--recursive"],
        cwd=repo_root,
        capture_output=True,
        text=True
    )
    if result.returncode == 0:
        print("Submodule initialized successfully.")
        if result.stdout:
            print(result.stdout)
    else:
        print("ERROR: Failed to initialize submodule.")
        print(result.stderr)
```
